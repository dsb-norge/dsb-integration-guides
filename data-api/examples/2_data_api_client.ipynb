# In this example
In this example we will use the `DsbDataApiClient` to fetch full datasets from the Data API and load them into a Pandas DataFrame.
<br>We will then demonstrate how to convert the DataFrame into different formats such as JSON, Parquet, and CSV.

```python
# --------------------------------------------------
# Get environment variables
# --------------------------------------------------
from dotenv import load_dotenv
load_dotenv(override=True, dotenv_path=".env-prod")
```

```python
# --------------------------------------------------
# Setup DsbDataApiClient
# --------------------------------------------------
from lib.dataapiclient import DsbDataApiClient

# Setup DsbDataApiClient
client = DsbDataApiClient()
```

```python
import logging
logging.basicConfig(level=logging.INFO)

df = client.get_dataset("finance_norway/fact_claim") \
    .top(100000) \
    .exclude("CompanyCode") \
    .collect() \
    .to_dataframe()
```

```python
df.info()
```

```python
import pandas as pd
df.to_csv(f"./data/finance_norway_fact_claim_{pd.Timestamp.now().strftime('%Y%m%d')}.csv", index=False)
```
