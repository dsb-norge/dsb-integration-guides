# In this example
In this example we will use the `DsbDataApiClient` to fetch full datasets from the Data API and load them into a Pandas DataFrame.
<br>This is not something you have to use, but it is a convenient way to work with the api as it abstracts away the complexities of pagination and retries.
<br>It also provide a simple interface for making use of the select, filter and sort features of the API.
<br>We will then demonstrate how to convert the DataFrame into different formats such as JSON, Parquet, and CSV.

```python
# --------------------------------------------------
# Get environment variables
# --------------------------------------------------
from dotenv import load_dotenv

load_dotenv(override=True, dotenv_path=".env")
```

```python
# --------------------------------------------------
# Setup DsbDataApiClient
# --------------------------------------------------
from lib.dataapiclient import DsbDataApiClient

# Setup DsbDataApiClient
client = DsbDataApiClient()
```

```python
# --------------------------------------------------
# Retrieve dataset
# --------------------------------------------------
df = (
    client.get_dataset("finance_norway/fact_claim")
    .page_size(100_000)
    .collect()
    .to_dataframe()
)
```

```python
# --------------------------------------------------
# Write info about the dataset
# --------------------------------------------------
df.info()
```

```python
import datetime

df.to_csv(
    f"./data/finance_norway_fact_claim_{datetime.date.today().strftime('%Y%m%d')}.csv",
    index=False,
)
```
